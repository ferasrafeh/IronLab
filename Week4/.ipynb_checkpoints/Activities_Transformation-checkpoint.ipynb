**Categorical columns:**

1. Region, Response, Month, Employment_status, Gender, Location_code, marital_satus, policy_type, sales_channel, and vehicle_class are of type NOMINAL. So, creating dummy variable would be the most suitable. (ONE-HOT ENCODING)
2. Coverage, education, policy, renew_offer_type, and vehicle_size are of type ORDINAL. So, ORDINAL ENCODING is suitable.

```python
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
```

## Creating Dummy Variables for the Nominal Columns

```python
cat = pd.get_dummies(cat,columns=["region","response","month","employment_status","gender","location_code",
                                  "marital_status","policy_type","sales_channel","vehicle_class"], dtype=int)

```

```python
cat.head()
```

## Ordinal Encoding for the ordinal columns

```python
 from sklearn.preprocessing import OrdinalEncoder
```

```python
cols = ['coverage', 'education', 'policy', 'renew_offer_type','vehicle_size']
oe = OrdinalEncoder()
cat[cols] = oe.fit_transform(cat[cols])
```

```python
cat
```

```python
# or to specify the heirarchy:
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(["high school or below","college","bachelor","master","doctor"])

cat["education"]= le.transform(cat["education"])

# repeat for other ordinal columns
```

## Begin by visually examining distributions (histograms) of the numerical features. Select a variable, call it varA, which takes on a wide range of numerical values, and another, varB, which has a noticeably large skew. For example, you might select customer_lifetime_value as a candidate varB which has skew.

```python
import seaborn as sns
import matplotlib.pyplot as plt

for column in num.columns:
    sns.displot(num[column], kde=True)
    plt.title(column)
```

```python
varA = num["income"].to_numpy().reshape(-1,1)
varB = num["customer_lifetime_value"].to_numpy().reshape(-1,1)
```

### varA:

### Use minmax transform to bring varA's values into the range [0,1].
### Check that varA has been rescaled using a displot or a histogram

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
varA_scaled = scaler.fit_transform(varA)
```

```python
# or
# scaler.fit(varA)
# scaler.transform(varA)
```

```python
sns.displot(varA_scaled)
plt.title("Income")
```

```python
num["income"] = varA_scaled
```

### varB

### Use StandardScaler or PowerTransform to reduce the skew and standardize varB's distribution.
### Check that the result has zero mean, unit variance, and reduced skew using mean(), std(), and a plot of the PDF.

```python
pt = PowerTransformer(method='box-cox')
varB_transformed = pt.fit_transform(varB)
scaler = StandardScaler()
varB_scaled = scaler.fit_transform(varB_transformed)
```

```python
print('Mean:', varB_scaled.mean())
print('Standard deviation:', varB_scaled.std())
sns.displot(varB_scaled)
plt.title("Customer Lifetime Value")
```

```python
num["customer_lifetime_value"] = varB_scaled
```
