```python
import pandas as pd

df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
numerical = df.select_dtypes("int64")
```

```python
categorical = df.select_dtypes("object")
```

```python
numerical.columns
```

```python
for column in numerical.columns:
    sns.displot(numerical[column], kde=True)
    plt.title(column)
    
```

```python
for column in numerical.columns:
    plt.hist(x = numerical[column])
    plt.title(column)
    plt.show()
```

### The distributions do not look like normal distributions

### Checking the correlation but not including the target feature "total_claim_amount".


```python
corr = numerical.iloc[:,0:7].corr()
corr
```

```python
matrix = np.triu(corr)
np.fill_diagonal(matrix, False)
sns.heatmap(corr, annot=True, fmt='.3f', mask=matrix)
```

#  <ins>**Activity 4:**</ins>

### - Plotting of the total number of responses

```python
responses_count = sns.countplot(x="response",data=categorical)

for p in responses_count.patches:
    responses_count.annotate(f'\n{p.get_height()}', 
                             (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='white', size=18)
```

### Plotting the response by sales channel

```python
responses_sales = sns.countplot(x="sales_channel",data=categorical, hue="response")
```

### Plotting response by total claim amount

```python
cutty = pd.cut(x=df["total_claim_amount"],bins=6)
```

```python
plt.figure(figsize=(12,10))
plt.title("Response by total claim amount")
responses_claim = sns.countplot(x=cutty,data=categorical,hue="response")
```

### Plotting response by income

```python
cutty_1 = pd.cut(x=df["income"],bins=5)
```

```python
plt.figure(figsize=(10,8))
plt.title("Response by Income")
responses_inc = sns.countplot(x=cutty_1,data=df, hue="response")
```
